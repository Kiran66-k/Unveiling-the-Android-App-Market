Sentiment Analysis

Assess user sentiments through reviews.

```python
import os
import pandas as pd

# Construct the full path to 'user_reviews.csv'
file_path_reviews = os.path.join(path, 'datasets', 'user_reviews.csv')

# Load the CSV file into a pandas DataFrame
user_reviews_df = pd.read_csv(file_path_reviews)

# Display the first 5 rows of the DataFrame
print("User reviews DataFrame loaded successfully. Displaying head:")
print(user_reviews_df.head())
```

```python
print("Data types of each column:")
user_reviews_df.info()

print("\nMissing values per column:")
print(user_reviews_df.isnull().sum())

print("\nDescriptive statistics for numerical columns:")
print(user_reviews_df.describe())
```

```python
print("Original shape of user_reviews_df:", user_reviews_df.shape)

# Drop rows where 'Translated_Review' is missing
user_reviews_df.dropna(subset=['Translated_Review'], inplace=True)

print("New shape of user_reviews_df after dropping nulls in 'Translated_Review':", user_reviews_df.shape)
print("\nVerifying no remaining missing values:")
print(user_reviews_df.isnull().sum())
```

```python
from textblob import TextBlob

# Function to get sentiment polarity and label
def get_sentiment(text):
    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity
    if polarity > 0:
        sentiment = 'Positive'
    elif polarity < 0:
        sentiment = 'Negative'
    else:
        sentiment = 'Neutral'
    return polarity, sentiment

# Apply sentiment analysis to the 'Translated_Review' column
user_reviews_df[['TextBlob_Polarity', 'TextBlob_Sentiment']] = user_reviews_df['Translated_Review'].apply(lambda text: pd.Series(get_sentiment(text)))

# Display the first few rows with the new sentiment columns
print("User reviews DataFrame with TextBlob sentiment analysis:")
print(user_reviews_df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the occurrences of each unique sentiment label
sentiment_counts = user_reviews_df['TextBlob_Sentiment'].value_counts()

# Create a bar chart to visualize the distribution
plt.figure(figsize=(10, 6))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, hue=sentiment_counts.index, palette='viridis', legend=False)
plt.title('Distribution of TextBlob Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
avg_sentiment_by_app = user_reviews_df.groupby('App')['TextBlob_Polarity'].mean().sort_values(ascending=False)
print("Top 10 Apps by Average TextBlob Sentiment Polarity:")
print(avg_sentiment_by_app.head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 7))
sns.barplot(x=avg_sentiment_by_app.head(10).index, y=avg_sentiment_by_app.head(10).values, hue=avg_sentiment_by_app.head(10).index, palette='viridis', legend=False)
plt.title('Top Apps by Average Sentiment Polarity')
plt.xlabel('App')
plt.ylabel('Average Sentiment Polarity')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

**Prepare data for interactive sentiment visualization**

```python
 app_sentiment_summary = user_reviews_df.groupby('App').agg(
    avg_sentiment=('TextBlob_Polarity', 'mean'),
    review_count=('Translated_Review', 'count')
).reset_index()

print("Head of app_sentiment_summary:")
print(app_sentiment_summary.head())
```
